```python
engorging_sum = pd.read_csv("/mnt/DATA2/zhong/BiteOScope_Data/An_Stephen_Vivax_Engorge_File/TrackStats_stephen_landing_vivax_fullstat.csv")
engorging_fraction_PF = engorging_sum.loc[engorging_sum['Infected'] == 0, 'Total engorging fraction'].values
engorging_fraction_C = engorging_sum.loc[engorging_sum['Infected'] == 1, 'Total engorging fraction'].values
engorging_fraction_PF
```

```python
landing_sum = pd.read_csv("/mnt/DATA2/zhong/BiteOScope_Data/An_Stephen_Vivax_Engorge_File/TrackStats_stephen_landing_vivax_fullstat.csv")
total_landing_PF = landing_sum.loc[landing_sum['Infected'] == 0, 'Total landing'].values
total_landing_C = landing_sum.loc[landing_sum['Infected'] == 1, 'Total landing'].values
total_landing_PF
```

```python
landing_wilcoxon = scipy.stats.ranksums(total_landing_PF,total_landing_C, alternative='two-sided', keepdims=False)
PF_stephen_landing_avg = np.average(total_landing_PF)
C_stephen_landing_avg = np.average(total_landing_C)
print(landing_wilcoxon)
print('PF_stephen_landing_avg = ', total_landing_PF)
print('C_stephen_landing_avg = ', total_landing_C)
```

```python
total_landing_percage_permos_PF = landing_sum.loc[landing_sum['Infected'] == 0, 'Total landing fraction'].values
total_landing_percage_permos_C = landing_sum.loc[landing_sum['Infected'] == 1, 'Total landing fraction'].values
total_landing_percage_permos_PF
```

```python
landing_percage_permos_wilcoxon = scipy.stats.ranksums(total_landing_percage_permos_PF,total_landing_percage_permos_C, alternative='two-sided', keepdims=False)
PF_stephen_landing_percage_permos_avg = np.average(total_landing_percage_permos_PF)
C_stephen_landing_percage_permos_avg = np.average(total_landing_percage_permos_C)
print(landing_percage_permos_wilcoxon)
print('PF_stephen_landing_percage_permos_avg = ', PF_stephen_landing_percage_permos_avg)
print('C_stephen_landing_percage_permos_avg = ', C_stephen_landing_percage_permos_avg)
```

  Not-Engorged File analysis  

```python
vivax_NoEngorge_0417 = pd.read_hdf('/mnt/DATA2/zhong/BiteOScope_Data/Result_Stat/20250417_An_Stephen_Vivax_NoEn_trackdata.h5')
vivax_NoEngorge_0418 = pd.read_hdf('/mnt/DATA2/zhong/BiteOScope_Data/Result_Stat/20250418_An_Stephen_Vivax_NoEn_trackdata.h5')
NoEngorge_Vivax_20250531 = pd.concat([vivax_NoEngorge_0417,vivax_NoEngorge_0418])
PF_NoEngorge_vivax = NoEngorge_Vivax_20250531[NoEngorge_Vivax_20250531['infected'] == 1]
C_NoEngorge_vivax = NoEngorge_Vivax_20250531[NoEngorge_Vivax_20250531['infected'] == 0]
```

```python
NoEngorge_totTime_wilcoxon_20250531 = scipy.stats.ranksums(C_NoEngorge_vivax['totTime'],PF_NoEngorge_vivax['totTime'], alternative='two-sided', keepdims=False)
NoEngorge_C_stephen_totTime_avg_20250531 = np.average(C_NoEngorge_vivax['totTime'])/25
NoEngorge_C_stephen_totTime_std_20250531 = np.std(C_NoEngorge_vivax['totTime'])/25
NoEngorge_PF_stephen_totTime_avg_20250531 = np.average(PF_NoEngorge_vivax['totTime'])/25
NoEngorge_PF_stephen_totTime_std_20250531 = np.std(PF_NoEngorge_vivax['totTime'])/25
print(NoEngorge_totTime_wilcoxon_20250531)
print('NoEngorge_C_stephen_totTime_avg_20250531 = ', NoEngorge_C_stephen_totTime_avg_20250531)
print('NoEngorge_C_stephen_totTime_std_20250531 = ', NoEngorge_C_stephen_totTime_std_20250531)
print('NoEngorge_PF_stephen_totTime_avg_20250531 = ', NoEngorge_PF_stephen_totTime_avg_20250531)
print('NoEngorge_PF_stephen_totTime_std_20250531 = ', NoEngorge_PF_stephen_totTime_std_20250531)
```

```python
NoEngorge_totDistSm_wilcoxon_20250531 = scipy.stats.ranksums(C_NoEngorge_vivax['totDistSm'],PF_NoEngorge_vivax['totDistSm'], alternative='two-sided', keepdims=False)
NoEngorge_C_stephen_totDistSm_avg_20250531 = np.average(C_NoEngorge_vivax['totDistSm'])*0.04
NoEngorge_C_stephen_totDistSm_std_20250531 = np.std(C_NoEngorge_vivax['totDistSm'])*0.04
NoEngorge_PF_stephen_totDistSm_avg_20250531 = np.average(PF_NoEngorge_vivax['totDistSm'])*0.04
NoEngorge_PF_stephen_totDistSm_std_20250531 = np.std(PF_NoEngorge_vivax['totDistSm'])*0.04
print(NoEngorge_totDistSm_wilcoxon_20250531)
print('NoEngorge_C_stephen_totDistSm_avg_20250531 = ', NoEngorge_C_stephen_totDistSm_avg_20250531)
print('NoEngorge_C_stephen_totDistSm_std_20250531 = ', NoEngorge_C_stephen_totDistSm_std_20250531)
print('NoEngorge_PF_stephen_totDistSm_avg_20250531 = ', NoEngorge_PF_stephen_totDistSm_avg_20250531)
print('NoEngorge_PF_stephen_totDistSm_std_20250531 = ', NoEngorge_PF_stephen_totDistSm_std_20250531)
```

```python
PF_NoEngorge_vivax['PF_speed'] = (PF_NoEngorge_vivax['totDistSm']*0.04).copy() / (PF_NoEngorge_vivax['totTime']/25).copy()
C_NoEngorge_vivax['C_speed'] = (C_NoEngorge_vivax['totDistSm']*0.04).copy() / (C_NoEngorge_vivax['totTime']/25).copy()
PF_NoEngorge_Mean_Speed = PF_NoEngorge_vivax['PF_speed'].values.copy()
C_NoEngorge_Mean_Speed = C_NoEngorge_vivax['C_speed'].values.copy()
NoEngorge_Mean_speed_wilcoxon_final = scipy.stats.ranksums(PF_NoEngorge_Mean_Speed,C_NoEngorge_Mean_Speed, alternative='two-sided', keepdims=False)
C_stephen_NoEngorge_Mean_speed_avg_final = np.average(C_NoEngorge_Mean_Speed)
C_stephen_NoEngorge_Mean_speed_std_final = np.std(C_NoEngorge_Mean_Speed)
PF_stephen_NoEngorge_Mean_speed_avg_final = np.average(PF_NoEngorge_Mean_Speed)
PF_stephen_NoEngorge_Mean_speed_std_final = np.std(PF_NoEngorge_Mean_Speed)
print(NoEngorge_Mean_speed_wilcoxon_final)
print('C_stephen_NoEngorge_Mean_speed_avg_final = ', C_stephen_NoEngorge_Mean_speed_avg_final)
print('C_stephen_NoEngorge_Mean_speed_std_final = ', C_stephen_NoEngorge_Mean_speed_std_final)
print('PF_stephen_NoEngorge_Mean_speed_avg_final = ', PF_stephen_NoEngorge_Mean_speed_avg_final)
print('PF_stephen_NoEngorge_Mean_speed_std_final = ', PF_stephen_NoEngorge_Mean_speed_std_final)
```

```python
NoEngorge_movingFraction_wilcoxon_final = scipy.stats.ranksums(C_NoEngorge_vivax['movingFrac'],PF_NoEngorge_vivax['movingFrac'], alternative='two-sided', keepdims=False)
C_stephen_NoEngorge_movingFraction_avg_final = np.average(C_NoEngorge_vivax['movingFrac'])
C_stephen_NoEngorge_movingFraction_std_final = np.std(C_NoEngorge_vivax['movingFrac'])
PF_stephen_NoEngorge_movingFraction_avg_final = np.average(PF_NoEngorge_vivax['movingFrac'])
PF_stephen_NoEngorge_movingFraction_std_final = np.std(PF_NoEngorge_vivax['movingFrac'])
print(NoEngorge_movingFraction_wilcoxon_final)
print('C_stephen_NoEngorge_movingFraction_avg_final = ', C_stephen_NoEngorge_movingFraction_avg_final)
print('C_stephen_NoEngorge_movingFraction_std_final = ', C_stephen_NoEngorge_movingFraction_std_final)
print('PF_stephen_NoEngorge_movingFraction_avg_final = ', PF_stephen_NoEngorge_movingFraction_avg_final)
print('PF_stephen_NoEngorge_movingFraction_std_final = ', PF_stephen_NoEngorge_movingFraction_std_final)
```
