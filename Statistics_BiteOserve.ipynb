Average probing bout - No engorge

```python
folder_AnSteph_Pv_1 = '/mnt/DATA2/zhong/BiteOServer/Data_Pf/labels/Analyzed_NoEngorge/20250417_An_Stephen_Vivax/Sequence_behavior'
folder_AnSteph_Pv_2 = '/mnt/DATA2/zhong/BiteOServer/Data_Pf/labels/Analyzed_NoEngorge/20250418_An_Stephen_Vivax/Sequence_behavior'
folder_AnSteph_Pv = [folder_AnSteph_Pv_1, folder_AnSteph_Pv_2]


result = []
for folder in folder_AnSteph_Pv:
    for root, dirs, files in os.walk(folder):
        for f in files:
            filepath = os.path.join(root, f)
            try:
                with open(filepath, 'r', encoding='utf-8') as file:
                    count = file.read().count('probe')
                    result.append([f, count])
            except:
                try:
                    with open(filepath, 'r', encoding='latin-1') as file:
                        count = file.read().count('probe')
                        result.append([f, count])
                except:
                    pass
df = pd.DataFrame(result, columns=['filename', 'probe_count'])
df.to_csv('/mnt/DATA2/zhong/BiteOServer/AnSteph_Pv_noengorge_probe_count_summary.csv', index=False)
```
